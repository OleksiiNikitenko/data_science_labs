```python
import pandas as pd
import seaborn
from matplotlib import pyplot as plt

data = pd.read_excel("rolling_sales/rollingsales_bronx.xls", skiprows=4)
data
#зчитуємо дані в районі Бронкс
```

```python
# далі очищуємо дані від зайвих пробілів, символів і так далі, щоб було зручніше працюати з даними 
clean_data = pd.read_excel("rolling_sales/rollingsales_bronx.xls", skiprows=4)
clean_data.columns = [column.lower().replace("\n", "").replace(" ", "_").replace("-", "_") for column in clean_data.columns] 
clean_data
```

```python
# Беремо building_class_category та розподіляємо всі будинки на самі основні категорії
building_categories = ["homes","condos", "apartments", "buildings", "residential", "factories", "garages", 
                       "warehouses", "land", "facilities"]

def add_build_category (category:str):
    buffer = ""
    for type in building_categories:
        if type.upper() in category:
            buffer = type
            break
    return buffer if buffer != "" else "other"
                                
                    
# створюємо новий стовбець type_building та додаємо його в дата фрейм         
clean_data.insert(2,"type_building",[add_build_category(category) for category in clean_data.building_class_category.to_list()])
new_data_with_type_building = clean_data
new_data_with_type_building
```

```python
# робимо групбай по тайп-білдінг та рахуємо іх кількість покупок
grouped_type_build_data = new_data_with_type_building.groupby("type_building",as_index=False).count()
grouped_type_build_data
```

```python
# малюємо графік та бачимо що більше всього було продамо HOMES, apartaments
fig, ax = plt.subplots(figsize=(20, 10))
ax = seaborn.barplot( data=grouped_type_build_data, 
                x=grouped_type_build_data.type_building, 
                y=grouped_type_build_data.borough)

ax.set(xlabel='Property type', ylabel='Quantity')
```

```python
# на графіку бачиму кількість продажів за 1 рік. З 2012.08 по 2013.08.
# Можемо зробити висновок, що пік продажів припадає на 2012.12, а спад продажів відбуваеться в кінці 2013.08
fig, group_data_by_year = plt.subplots(figsize=(10,5))
group_data_by_year = clean_data.groupby([clean_data["sale_date"].dt.year, clean_data["sale_date"].dt.month],group_keys=True)["saleprice"].sum()
group_data_by_year.plot(xlabel="Year and month", ylabel="Sum of sale price")
```

```python
# робимо графік по ціні за виду будинку. Бачимо що найдорожчий вид будинку - це апартаменти, та в два рази дешевші ХОМС
fig, ax = plt.subplots(figsize=(20, 10))
group2 = clean_data.groupby(["type_building"],as_index=False).sum()
ax = seaborn.barplot(data=group2, 
               x=group2.type_building, 
               y=group2.saleprice)
ax.set(xlabel='Property type', ylabel='Sum of sale price')
```
